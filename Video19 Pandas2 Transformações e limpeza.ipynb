```python
# Transformacoes com APPLY, MAP e APPLYMAP
import pandas as pd
import numpy as np
dtips = pd.read_csv('tips.csv') # importar com o pandas cria automaticamente um DF
```

Apply - aplica uma funcao ao longo das linhas (axis =0)ou colunas (axis=1) de um DF
MAP - aplica uma funcao em cada elemento de uma **Serie**
Applymap - aplica uma funcao em cada elemento de um **DF**


```python
dtips
```

```python
# Calular a soma das contas e gorgetas ao longo das linhas. Faz uma soma total
dtips[['total_bill','tip']].apply(np.sum)

```

```python
# Calcular a soma da conta e gorgeta ao longo das colunas (total do dia)
# agrupando por dia e trazendo o resultado em uma coluna
dtips[['total_bill','tip']].apply(np.sum, axis=1)
```

```python
# criando uma coluna com gorjeta media por pessoa na mesa
dtips['tip_perperson'] = dtips.apply(lambda x: np.round(x['tip']/x['size'],2), axis =1)
dtips

# nao entendi o x:
```

```python
#Utilizando o Applymap - funciona em 2 colunas. Aplicado a um dataframe

dtips[['total_bill','tip']].applymap(np.round)
```

```python
# Utilizando o MAP - Nao aceita ser aplicado para 2 colunas ao mesmo tempo

dtips['total_bill'].map(np.round)
```

LIMPEZA DE DADOS - TRATANDO COMO FALTANTES E DUPLICADOS


```python
dtips = pd.read_csv('tips.csv') # importar com o pandas cria automaticamente um DF
pd.isna(dtips) # Verifica se existe um Not a number no DF

# Como verificar se existe entre as colunas 5 e 239, que nao estao sendo exibidas

np.any(pd.isna(dtips)) # retorna true se a consicao for verdadeira
```

```python
# Percorrendo colunas - por padrao a iteracao eh por linhas, vamos fazer por colunas e procurar dados faltantes
#professor utilizou o metodo iteritems() mas foi removido do pandas, substitui por items()
import pandas as pd
print('Dados faltantes em:')
for (colName, colData) in dtips.items():
  if(np.any(pd.isna(colData))):
    print(colName)
```

```python
# Verificando not a number nas colunas
np.any(dtips.isna(), axis=1)
```

```python
# para tentar achar os nan, podemos usar o loc
dtips.loc[np.any(dtips.isna(), axis=1)]
```

```python
print(dtips.shape)
```

```python
# Comando dropna
# Remove as linhas e colunas onde dados faltantes existem
dtips_na_rows = dtips.dropna()
print(dtips_na_rows)
print(dtips_na_rows.shape)
```

```python
# Removendo as colunas onde tem NAN
dtips_na_rows = dtips.dropna(axis=1)
print(dtips_na_rows)
```

```python
# Outra alternativa eh preencher o NAN com algum valor. Para isso utilizamos o comando fillna
dtips.loc[np.any(dtips.isna(), axis=1)]
dtips_fill1 = dtips.fillna(0) # onde tinha NAN estou preenchendo com zero
dtips_fill1.loc[np.any(dtips.isna(),axis=1)]
```

```python
# realizando o preenchimento com dicionario

# crio o dicionario, vou querer preencher uma coluna com 10 e outra com 5
filldic = {'total_bill':10, 'tip':5}

#criando um novo DF e utilizando o fillna para preencher o DF com o dicionario criado
dtips_fill2 = dtips.fillna(filldic)

# Localizando se ha valores NA no DF
dtips_fill2.loc[np.any(dtips.isna(),axis=1)]

```

```python
# Tratando linhas duplicadas - comando duplicated e drop_duplicates()
print(dtips.duplicated())
# nao vejo se ha algo entre o 5 e o 239, pois as linhas estao colapsadas
# para resolver isso

print('\n', dtips.duplicated().any())
```

```python
# parametro KEEP
# False: retorna True para todas as duplicadas
# first: a primeira nao eh detectada
# last: a ultima nao eh detectada

dtips[dtips.duplicated(keep=False)]
```

```python
# mostrando as segundas linhas duplicadas
dtips[dtips.duplicated(keep='first')]
```

```python
# removendo as duplicadas com o drop duplicates
dtips_first = dtips.drop_duplicates(keep='first')
print(dtips_first.iloc[216:225])
# veja que as linhas 218 e 225 foram removidas
```

```python
# Apos a remocao, resete os indices para ajustar os dados
dtips_first = dtips_first.reset_index()
#verificando
print(dtips_first.iloc[216:225])
```
