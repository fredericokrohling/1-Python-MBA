```python
#calculo imc
altura = float(input('Digite a altura: '))
peso = float(input('Digite o peso: '))
print('O IMC é:', round(peso/altura**2,2))
```

```python
#Calculo deslocamento
acel = float(input('Digite a aceleracao do corpo: '))
vel_init = float(input('Digite a velocidade inicial: '))
tempo = float(input('Digite o intervalo de tempo: '))
s = vel_init*tempo + (acel*tempo**2/2)
print('O deslocamento é de %.2f' % s)
```
